```python
from fastai.tabular import *
```

```python
df = pd.read_csv('../input/forest-forest-dataset/forestfires.csv')
dep_var = 'area'
cat_names = ['month', 'day']
cont_names = ['X', 'Y', 'FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind','rain']
procs = [FillMissing, Categorify, Normalize]
```

```python
df.head()
```

```python
df.info()
```

```python
data = (TabularList.from_df(df, cat_names=cat_names, cont_names=cont_names, procs=procs)
                           .split_by_idx(valid_idx=range(413,516))
                           .label_from_df(cols=dep_var)
                           .databunch())
```

```python
data.show_batch()
```

```python
learn = tabular_learner(data, layers=[200,100], metrics=rmse)
learn.fit(5, 1e-2)
```

```python
learn.show_results()
```
